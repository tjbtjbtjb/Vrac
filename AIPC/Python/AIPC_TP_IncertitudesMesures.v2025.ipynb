## INTRODUCTION

Lors d'une mesure √©lectrique utilisant un multim√®tre (en tant que voltm√®tre, amp√®rem√®tre ou ohmm√®tre), l'√©cran de l'appareil affiche une seule valeur $m$. 

La valeur mesur√©e est naturellement cette valeur affich√©e. *Mais que vaut l'incertitude-type associ√©e ?*

On sait bien qu'**un r√©sultat de mesure est un ensemble (a priori infini) de valeurs raisonnablement attribuables au mesurande**, et que **l'incertitude-type est (une estimation de) l'√©cart-type de cet ensemble**. Il faut donc reconstituer cet ensemble. Comment faire ?

Les premiers √©l√©ments de r√©ponse sont √† chercher ‚Äî syst√©matiquement ‚Äî dans la notice de l'appareil. On y lit typiquement que ¬´ la pr√©cision est (0,2 % de la valeur lue + 3 fois le dernier digit)¬ª. Le fabricant caract√©rise ainsi, respectivement, le d√©faut d'√©talonnage de l'appareil et les erreurs al√©atoires ; il est capable de produire ces informations par l'analyse du grand nombre d'appareils identiques qu'il produit √† l'aide d'un syst√®me d'√©talonnage performant.

Pour d√©crire le **r√©sultat de la mesure**, il est d'usage de consid√©rer que la *pr√©cision* Œî d√©clar√©e par le fabricant correspond l'ensemble des valeurs contenue dans l'intervalle $[m - Œî, m + Œî]$, et √† supposer que toutes les valeurs sont √©quiprobables. Autrement dit, le r√©sultat de mesure est d√©crit par une distribution de probabilit√© uniforme sur l'intervalle $[m - Œî, m + Œî]$. 

Dans ce cas, l'**expression du r√©sultat de mesure est donc** :
- la valeur mesur√©e m
- l'incertitude-type $u(m) = \Delta/\sqrt{3}$ (r√©sultat classique pour l'√©cart-type d'une distribution uniforme de demi-√©tendue Œî).

Parfois, on **exprime le r√©sultat de la mesure** sous la forme $m ¬± u(m)$, en pr√©cisant que ce qui suit le ¬± est l'incertitude-type. Par exemple, on peut annoncer le r√©sultat d'une mesure de tension sous la forme $U = (13,07 ¬± 0,10)$ V. On peut √©galement √©crire : \
U = 13,07 V \
u(U) = 0,10  V

*Remarques :*
- Ce sont des r√®gles semblables qui s'appliquent pour les valeurs nominales des r√©sistances de r√©sistors. Par exemple, une r√©sistance de "100 Œ© ¬± 5%" d'apr√®s son fabricant correspond √† $R = (100,0 ¬± 2,9)$ Œ© (ce qui suit le signe ¬± est l'incertitude-type).
- De mani√®re g√©n√©rale, l'√©valuation  de l'incertitude √† l'aide d'une indication unique et en faisant des hypoth√®ses raisonnables pour estimer un √©cart-type, est d√©nomm√©e *√©valuation de type B*.

### üìù Mini-fiche : R√©gression affine en Python

La r√©gression affine sert √† ajuster un mod√®le : $y = a x + b$

Elle estime la pente **a** et l'ordonn√©e √† l'origine **b** √† partir de
donn√©es. Elle suppose que :
- $x$ est connu sans incertitude (ou ce qui revient au m√™me avec une incertitude n√©gligeable devant les autres)
- $y$ est connue avec une incertitude, inconnue, mais celle-ci est la m√™me pour toutes les valeurs de $y$.


``` python

x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 5, 4])

a, b = np.polyfit(x, y, 1)   # 1 ‚Üí degr√© 1 (droite)
print(a, b)
```

*Remarques* : 
- la r√©gression affine n'est pas la m√™me chose qu'une r√©gression lin√©aire !
- on n'a pas l'incertitude-type associ√©e aux param√®tres : la r√©cup√©rer est faisable, mais pas √©vident (simulation Monte Carlo)

### üìù Mini-fiche : R√©gression lin√©aire en Python

La r√©gression lin√©aire sert √† ajuster un mod√®le : $y = a x$

Elle estime la pente **a**. 

Il y a 2 mani√®res de proc√©der.

1) On fait **comme dans le secondaire**

    On calcule les rapports $y_i/x_i$, on obtient une s√©rie de valeurs $a_i$. On en calcule la moyenne arithm√©tique et on fait une √©valuation de type A de l'incertitude-type

``` python

x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 5, 4])

val_a = np.average(y/x)
u_a = np.std(y/x, ddof=1) / np.sqrt(len(a))
print(val_a, u_a)
```

_Attention :_ cette proc√©dure suppose que la loi lin√©aire est bien v√©rifi√©e.

2) On utilise une **m√©thode avanc√©e**

    Il est possible de montrer que sous l'hypoth√®se que les incertitudes-types sur $x$ sont nulles (ou n√©gligeables) et que celles sur $y$ sont identiques (et √©gales √† u(y)), alors dans un certain sens math√©matique, la ¬´ meilleure ¬ª estimation de $a$ est donn√©e par :

    $a = \dfrac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2}$

    et son incertitude-type vaut : 

    $u(a) = \dfrac{u(y)}{\sqrt{\sum_{i=1}^n x_i^2}}$
    
``` python

x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 5, 4])
```

### üìù Mini-fiche : √âvaluation d'incertitude des param√®tres d'une r√©gression affine par m√©thode de Monte Carlo

Lire la cellule suivante...

````python
x = np.array([1, 2, 3, 4]) # valeurs arbitraires
y = np.array([2, 3, 4.5, 4.5]) # valeurs arbitraires
u_x = x/20  # incertitude-type arbitraire
u_y = y/10  # incertitude-type arbitraire
n = len(x)

NMC = 100000 # nombre de simulations Monte Carlo

a = np.zeros(NMC) # tableau pour stocker les r√©sultats des pentes
b = np.zeros(NMC) # tableau pour stocker les r√©sultats des ordonn√©es √† l'origine

for i in range(NMC): # boucle de simulations Monte Carlo
    x_sim = x + np.random.normal(0, u_x, n) # simulation des x
    y_sim = y + np.random.normal(0, u_y, n) # simulation des y
    a[i], b[i] = np.polyfit(x_sim, y_sim, 1) # ajustement affine pour la simulation

val_a = np.average(a) # valeur moyenne des pentes simul√©es
u_a = np.std(a, ddof=1) # √©cart-type des pentes simul√©es
val_b = np.average(b) # valeur moyenne des ordonn√©es √† l'origine simul√©es
u_b = np.std(b, ddof=1) # √©cart-type des ordonn√©es √† l'origine simul√©es

print(f"a = {val_a:.2f} ¬± {u_a:.2f}")  # affichage des r√©sultats
print(f"b = {val_b:.2f} ¬± {u_b:.2f}")

plt.figure()
plt.errorbar(x, y, xerr=u_x, yerr=u_y, fmt='o', label='Donn√©es avec incertitudes-types') # trac√© des donn√©es avec barres d'erreur
x_fit = np.linspace(0, 5, 100) # intervalle de valeurs x pour la droite ajust√©e
y_fit = val_a * x_fit + val_b # calcul des y correspondants pour la droite ajust√©e
plt.plot(x_fit, y_fit, 'r-', label='Ajustement affine')  # trac√© de la droite ajust√©e
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()  

plt.figure()
plt.hist(a, bins='rice', density=True) # histogramme de densit√© des pentes simul√©es
plt.title('Distribution des pentes (a)')
plt.show()  
````
