```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z=np.loadtxt("20260107024801.csv",skiprows=22,delimiter=",")
```

```python
data=z[:,1]                  # values
t=z[:,0]                     # the time 
N=len(t)                     # Nb of samples
F_ech=1/(t[1]-t[0])          # sampling period
freq=np.arange(N)/N*F_ech    # freq in the Fourier space
```

```python
plt.plot(t,data)
plt.xlabel('time (in s)')
plt.ylabel('U (in V)')
plt.grid()
```

```python
ft=np.fft.fft(data)
i_max=300
plt.semilogy(freq[0:i_max],np.abs(ft[0:i_max]))
plt.xlabel('Frequency (in Hz)')
plt.ylabel('FFT amplitude')
plt.grid()
```
